```python
import pandas as pd
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium import webdriver
from bs4 import BeautifulSoup as bs
```

```python
driver = uc.Chrome()
url = 'https://finance.naver.com/sise/lastsearch2.naver'
driver.get(url)
```

```python
html = driver.page_source
```

```python
print(html)
```

```python
soup = bs(html)
type_5 = soup.find('table', {'class':'type_5'})
type_5
```

```python
tr_list = type_5.find_all('tr')
print(tr_list[2])
print(tr_list[3])
```

```python
for tr_index in tr_list :
    td_list = tr_index.find_all('td')
    print('----------')
    for td_index in td_list:
        print(td_index.text.strip())
```

```python
Data = pd.DataFrame(columns=['순위','종목명','검색비율','현재가',
                             '전일비','등락률','거래량','시가','고가','저가','PER','ROE'])
Data_num = 0
for tr_index in tr_list :
    td_list = tr_index.find_all('td')
    print('----------')
    data_list = []
    for td_index in td_list:
        print(td_index.text.strip())
        data_list.append(td_index.text.strip())
    try:
        Data.loc[Data_num] = data_list
    except:
        continue
    Data_num += 1
```

```python
print(Data.info())
```

```python
Data['순위'] = Data['순위'].astype('int')
Data[['PER', 'ROE']] = Data[['PER', 'ROE']].replace('N/A',None, regex=True).astype('float')
Data[['검색비율', '등락률']] = Data[['검색비율', '등락률']].replace('%','', regex=True).astype('float')
Data[['현재가', '전일비', '거래량', '시가', '고가', '저가']] = Data[['현재가', '전일비', '거래량', '시가', '고가', '저가']].replace(',', '', regex=True).astype('int')
```

```python
print(Data.info())
Data.to_excel('네이버증권_검색상위종목.xlsx', index=False)
```
